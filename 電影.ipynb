```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = ["https://movies.yahoo.com.tw/movieinfo_review.html/id=6058?sort=update_ts&order=desc&page={}".format(str(i)) for i in range(1,49)]
#改id數字和range裡的數值
```

```python
def yahooMovieParser(url):
    r = requests.get(url, headers={'Connection':'close'})
    soup = BeautifulSoup(r.text,'lxml')
    tags = soup.find_all("a", class_= "btn_report")
    for a in tags: 
        print(a.find_next('span').text)
```

```python
MovieInfo = None
 
for u in url:
    d1 = yahooMovieParser(u)
    if MovieInfo is None:
        MovieInfo = d1
    else:
        MovieInfo = False
```
